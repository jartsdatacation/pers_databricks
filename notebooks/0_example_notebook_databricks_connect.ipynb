### 0.1 Obtain parameters and secrets

Parameters are defined as widgets in the notebook. This allows you to run the notebook with different parameters without changing the code. Task (and job) parameters are retrieved via widgets, which makes it easy to parameterize the notebook for deployment in a Databricks Job.

Secrets are stored in a Databricks secret scope. This allows you to securely store sensitive information, such as API keys, without hardcoding them in the notebook.

```python
from databricks.sdk.runtime import dbutils


if "environment" in dbutils.widgets._widgets:
    ENVIRONMENT = dbutils.widgets.get("environment")
else:
    ENVIRONMENT = dbutils.widgets.text("environment", "dev")
if "uc_catalog" in dbutils.widgets._widgets:
    UC_CATALOG_SOURCE = dbutils.widgets.get("uc_catalog")
else:
    UC_CATALOG_SOURCE = dbutils.widgets.text("uc_catalog", "dev")
try:
    UC_SCHEMA_SOURCE = dbutils.widgets.get("uc_schema")
except KeyError:
    UC_SCHEMA_SOURCE = dbutils.widgets.text("uc_schema", "00_bronze")

EXAMPLE_SECRET = dbutils.secrets.get(scope=f"{ENVIRONMENT}-secrets-scope", key="example-secret")
```
